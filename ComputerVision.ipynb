```python
from sklearn.datasets import load_digits
```

```python
dt = load_digits(n_class=3)
```

```python
images = dt.images
```

```python
Y = dt.target
```

```python
images[-2]
```

```python
Y[-2]
```

```python
images[0].reshape(1, 64)
```

```python
images.shape
```

```python
X = images.reshape(-1,64)
```

```python
X.shape
```

```python
X = torch.FloatTensor(X)
Y = torch.LongTensor(Y)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
model = nn.Sequential(
    nn.Linear(64, 32),
    nn.ReLU(),
    nn.Linear(32, 3),
)
```

```python
# model(X)
```

```python
# nn.NLLLoss()
```

```python
lossfn = nn.CrossEntropyLoss()
```

```python
opt = optim.Adam(model.parameters(), lr=0.01)
```

```python
for _ in range(30):
    opt.zero_grad()
    Yp = model(X)
    loss = lossfn(Yp, Y)
    loss.backward()
    opt.step()
    print(loss)
```
